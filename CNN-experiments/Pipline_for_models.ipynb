```python
import tensorflow as tf
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import clear_output
clear_output()
```

### Цвет

```python
model_color = tf.keras.models.load_model("./models/b-cnn_only_color_v1.1_76.79%")
```

```python
multicolor = "./data/2DogCategories_color_and_tail/color/only_multicolored/138.jpg"
dark = "./data/2DogCategories_color_and_tail/color/only_dark/163.jpg"
light = "./data/2DogCategories_color_and_tail/color/only_light/2381.jpg"
m_img = cv2.imread(multicolor)
d_img = cv2.imread(dark)
l_img = cv2.imread(light)
plt.figure(figsize=(16,8))
plt.subplot(1, 3, 1)
plt.imshow(m_img)
plt.subplot(1, 3, 2)
plt.imshow(d_img)
plt.subplot(1, 3, 3)
plt.imshow(l_img)
```

```python
m_img = cv2.resize(m_img, (60, 70))
new_m_img = m_img.reshape(1, 60, 70, 3) / 255

d_img = cv2.resize(d_img, (60, 70))
new_d_img = d_img.reshape(1, 60, 70, 3) / 255

l_img = cv2.resize(l_img, (60, 70))
new_l_img = l_img.reshape(1, 60, 70, 3) / 255
```

```python
M_pred = model_color.predict(new_m_img)
m_pred = np.argmax(M_pred, axis=1)

D_pred = model_color.predict(new_d_img)
d_pred = np.argmax(D_pred, axis=1)

L_pred = model_color.predict(new_l_img)
l_pred = np.argmax(L_pred, axis=1)

print(m_pred, d_pred, l_pred)
```

Мультицветная, темная и светлая соответственно

### Хвост

```python
model_tail = tf.keras.models.load_model("./models/b-cnn_only_tails_v1.2_64.29%")
```

```python
long_path = "./data/2DogCategories_color_and_tail/tail/only_long_tail/163.jpg"
short_path = "./data/2DogCategories_color_and_tail/tail/only_short_tail/138.jpg"
long_img = cv2.imread(long_path)
short_img = cv2.imread(short_path)
plt.figure(figsize=(16,8))
plt.subplot(1, 2, 1)
plt.imshow(long_img)
plt.subplot(1, 2, 2)
plt.imshow(short_img)
```

```python
long_img = cv2.resize(long_img, (60, 70))
new_long_img = long_img.reshape(1, 60, 70, 3) / 255

short_img = cv2.resize(short_img, (60, 70))
new_short_img = short_img.reshape(1, 60, 70, 3) / 255
```

```python
Long_pred = model_tail.predict(new_long_img)
long_pred = np.argmax(Long_pred, axis=1)

Short_pred = model_tail.predict(new_short_img)
short_pred = np.argmax(Short_pred, axis=1)

print(long_pred, short_pred)
```

Длинный и короткий хвост соответственно
